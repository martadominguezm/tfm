```python
G = nx.DiGraph()

G.add_nodes_from(probabilityMatrix_user2.keys())

for route1 in probabilityMatrix_user2:
    for route2, prob in probabilityMatrix_user2[route1].items():
        if prob > 0:
            G.add_edge(route1, route2, weight=prob)

edge_weights = [G[u][v]['weight'] * 5 for u, v in G.edges()]

pos = nx.circular_layout(G) 

plt.figure(figsize=(15, 12))
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightgrey', font_size=9.5, font_weight='bold', width=edge_weights, edge_color='orange', arrowsize=12)
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): f"{G[u][v]['weight']:.2f}" for u, v in G.edges()}, font_size=8, font_color='black')
plt.title("Grafo de Probabilidades de Transición - Usuario 2")
plt.show()
```
